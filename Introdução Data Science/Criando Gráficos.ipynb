```python
import pandas as pd
import matplotlib.pyplot as plt

# Não é a planilha de dados oficial do curso (eles não colocaram pra download), mas eu conseguir uma semelhante para estudar.
df = pd.read_csv('base de dados/netflix.csv')
df = df.drop(df.columns[0], axis=1)
```

```python
eixo_x = df['listed_in'].value_counts() # original era: .keys()
eixo_y = df['listed_in'].value_counts() # original era: .values

# adaptado apenas para mostrar 5 gêneros ao invés de todos

top5 = eixo_x.head(5) 
top5 = eixo_y.head(5)

eixo_x = top5.keys()
eixo_y = top5.values
```

```python
plt.bar(eixo_x, eixo_y)

plt.title('Filmes por gênero')
plt.xlabel('Gêneros de filme')
plt.ylabel('Quantidade de filmes')
plt.xticks(rotation=45)

plt.show()
```

```python
eixo_x = df['release_year'].value_counts() # original era:  .sort_index().keys()
eixo_y = df['release_year'].value_counts() # original era: .sort_index().values

#adaptado apenas para mostrar 5 gêneros ao invés de todos

top5 = eixo_x.head(5).sort_index()
top5 = eixo_y.head(5).sort_index()

eixo_x = top5.keys()
eixo_y = top5.values
```

```python
plt.plot(eixo_x, eixo_y)

plt.title('Melhores filmes por ano de lançamento')
plt.xlabel('Ano de lançamento')
plt.ylabel('Quantidade de filmes')

plt.show()
```
